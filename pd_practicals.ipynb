## Read YAML with pandas

```python
#Create a function that transforms a yaml file into a pandas DataFrame
def to_df(path):
    with open(path) as f:
        data = yaml.safe_load(f)
    return pd.json_normalize(data)

```

## Fire Fighters versus Police Officers
- Ratio between people in each department
- Mean salary of each department

```python
police_employees = sf_sal.loc[sf_sal["JobTitle"].str.contains("POLICE DEPARTMENT")]
fire_employees = sf_sal.loc[sf_sal["JobTitle"].str.contains("FIRE DEPARTMENT")]

police_count = police_employees['JobTitle'].count()
fire_count = fire_employees['JobTitle'].count()
print(f"The ratio of people in the fire department over people in the police department is Police: {police_count}, Fire: {fire_count}")

p_salary_mean = police_employees["BasePay"].mean().round(decimals = 2)
f_salary_mean = fire_employees["BasePay"].mean().round(decimals = 2)
print(f"The mean salary of the police department is {p_salary_mean}.")
print(f"The mean salary of the fire department is {f_salary_mean}.")


fire_employees.head(2)



```

## DataFrame Manipulation

```python
#Unique jobs
unique = sf_sal["JobTitle"].unique()
len(unique)
```

```python
#How many people have first name John
johns = sf_sal.loc[sf_sal["EmployeeName"].str.startswith("JOHN")]
johns["EmployeeName"].count()
```

```python
#Find the surname of everyone whose first name is more than 6 characters long

#Create first name and last name columns
sf_copy = sf_sal.copy()
employees = sf_copy.loc[:,["Id", "EmployeeName"]]

first_names = []
last_names = []

for i in employees["EmployeeName"]:
    split = i.split()
    first_names.append(split[0])
    last_names.append(split[1])
    
first_names = pd.DataFrame(first_names, index=None)
last_names = pd.DataFrame(last_names, index=None)
sf_copy["FirstName"] = first_names
sf_copy["Surname"] = last_names

sf_copy.head(5)
    
    
```

```python
six_char_employees = sf_copy.loc[sf_copy["FirstName"].str.len() > 6]
six_char_employees["Surname"]
```

```python
#Create a new column called "last_updated" where every value is equal to the current date and time (use the datetime module and format the date in ISO format)
from datetime import datetime as dt
now = dt.now()
sf_copy["last_updated"] = now.strftime("%d/%m/%Y %H:%M:%S")
sf_copy.head(3)

```

```python
#5.create a new derived column called "time_ratio" in the DataFrame which is the proportion of overtime pay to base pay
sf_copy["time_ratio"] = None 

```

```python
#In one line create a new DataFrame called new_df that only contains employees that make more than 100k base salary
new_df = sf_copy.loc[sf_copy["BasePay"] > 100000]
new_df.head(3)
```

```python
fn = lambda row: row.TotalPay - row.BasePay
col = new_df.apply(fn, axis=1)
new_df = new_df.assign(pay_minus_base=col.values)
new_df.head(3)
```

```python
total_pay = new_df["TotalPay"]
base_pay = new_df["BasePay"]
new_df["pay_without_base"] = total_pay - base_pay
new_df.head(3)
```
