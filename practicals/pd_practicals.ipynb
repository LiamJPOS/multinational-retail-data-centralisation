## Read YAML with pandas

```python
#Create a function that transforms a yaml file into a pandas DataFrame
def to_df(path):
    with open(path) as f:
        data = yaml.safe_load(f)
    return pd.json_normalize(data)

```

## Fire Fighters versus Police Officers
- Ratio between people in each department
- Mean salary of each department

```python
police_employees = sf_sal.loc[sf_sal["JobTitle"].str.contains("POLICE DEPARTMENT")]
fire_employees = sf_sal.loc[sf_sal["JobTitle"].str.contains("FIRE DEPARTMENT")]

police_count = police_employees['JobTitle'].count()
fire_count = fire_employees['JobTitle'].count()
print(f"The ratio of people in the fire department over people in the police department is Police: {police_count}, Fire: {fire_count}")

p_salary_mean = police_employees["BasePay"].mean().round(decimals = 2)
f_salary_mean = fire_employees["BasePay"].mean().round(decimals = 2)
print(f"The mean salary of the police department is {p_salary_mean}.")
print(f"The mean salary of the fire department is {f_salary_mean}.")


fire_employees.head(2)



```

```python
flights_df['AIRLINECODE_ASINT'] = flights_df['AIRLINECODE'].cat.codes.astype('int64')
flights_df['AIRLINECODE_ASINT']
```

```python
cdf = pd.read_csv('car_insurance.csv')
cdf
```

```python
cdf.info()
```

```python
#Missing values per column
for col in cdf:
    print(col, cdf[col].isna().sum())
```

```python
cdf = cdf.dropna()
cdf
```

```python
#Means
means = {'Age':cdf['Age'].mean(),
         'Balance':cdf['Balance'].mean(),
         'LastContactDay' : cdf['LastContactDay'].mean(),
         'DaysPassed': cdf['DaysPassed'].mean(),
         'PrevAtttempts': cdf['PrevAttempts'].mean()}

means

```

```python
#Medians
medians = {'Age':cdf['Age'].median(),
         'Balance':cdf['Balance'].median(),
         'LastContactDay' : cdf['LastContactDay'].median(),
         'DaysPassed': cdf['DaysPassed'].median(),
         'PrevAtttempts': cdf['PrevAttempts'].median()}

medians
```

```python
#Upper quartiles
quartiles = {'Age':cdf['Age'].quantile(q=0.75),
         'Balance':cdf['Balance'].quantile(q=0.75),
         'LastContactDay' : cdf['LastContactDay'].quantile(q=0.75),
         'DaysPassed': cdf['DaysPassed'].quantile(q=0.75),
         'PrevAtttempts': cdf['PrevAttempts'].quantile(q=0.75)}

quartiles
```
